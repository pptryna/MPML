```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Mengonversi data kategorikal menjadi numerik
label_encoders = {}
categorical_features = ['Gender', 'Marital Status', 'Occupation', 'Educational Qualifications','Monthly Income']

for feature in categorical_features:
    le = LabelEncoder()
    food[feature] = le.fit_transform(food[feature])
    label_encoders[feature] = le

```

```python
# Menentukan fitur dan label
X = food[['Age', 'Gender', 'Marital Status', 'Occupation', 'Monthly Income', 'Educational Qualifications', 'Family size']]
y = food['Feedback']

```

```python
print(X)
```

```python
# Normalisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
# Membagi data menjadi set pelatihan dan set pengujian
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

PEMODELAN

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC, SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
```

```python
# Membuat list objek model
models = [LogisticRegression(), LinearSVC(), SVC(kernel='rbf'), KNeighborsClassifier(),
          DecisionTreeClassifier(), GaussianNB(),RandomForestClassifier(n_estimators=200)]

# Daftar nama model yang sesuai
model_names = ['LogisticRegression', 'LinearSVM', 'rbfSVM', 'KNearestNeighbors',
               'DecisionTree', 'GaussianNB','RandomForest']
```

```python
# Melatih dan menguji setiap model
results = []
for model, model_name in zip(models, model_names):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    results.append({'Model': model_name, 'Accuracy': accuracy})

# Membuat DataFrame dari hasil
results_df = pd.DataFrame(results)

# Menampilkan tabel hasil
print(results_df)
```

MELAKUKAN HYERPARAMETER TUNING

```python
#SVM
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Inisialisasi model SVM
svm_model = SVC(kernel='linear')

# Tentukan parameter grid untuk tuning
param_grid_svm = {
    'C': [0.1, 1, 10, 100],  # Regularization parameter
}

# Inisialisasi GridSearchCV
grid_search_svm = GridSearchCV(estimator=svm_model, param_grid=param_grid_svm, cv=5, scoring='accuracy', n_jobs=-1)

# Fit model ke data
grid_search_svm.fit(X_train, y_train)

# Menampilkan hasil terbaik
print(f"Best parameters for SVM: {grid_search_svm.best_params_}")
print(f"Best cross-validation accuracy for SVM: {grid_search_svm.best_score_}")
```

```python
from sklearn.linear_model import LogisticRegression

# Inisialisasi model Logistic Regression
logreg_model = LogisticRegression(max_iter=10000)

# Tentukan parameter grid untuk tuning
param_grid_logreg = {
    'C': [0.1, 1, 10, 100],  # Regularization parameter
    'penalty': ['l1', 'l2'],  # Regularization technique
    'solver': ['liblinear', 'saga']  # Solvers that support L1 regularization
}

# Inisialisasi GridSearchCV
grid_search_logreg = GridSearchCV(estimator=logreg_model, param_grid=param_grid_logreg, cv=5, scoring='accuracy', n_jobs=-1)

# Fit model ke data
grid_search_logreg.fit(X_train, y_train)

# Menampilkan hasil terbaik
print(f"Best parameters for Logistic Regression: {grid_search_logreg.best_params_}")
print(f"Best cross-validation accuracy for Logistic Regression: {grid_search_logreg.best_score_}")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Inisialisasi model Random Forest
rf_model = RandomForestClassifier()

# Tentukan parameter grid untuk tuning
param_grid_rf = {
    'n_estimators': [100, 200, 300],  # Number of trees
    'max_features': ['auto', 'sqrt', 'log2'],  # Number of features to consider at each split
    'max_depth': [None, 10, 20, 30],  # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],  # Minimum number of samples required to be at a leaf node
}

# Inisialisasi GridSearchCV
grid_search_rf = GridSearchCV(estimator=rf_model, param_grid=param_grid_rf, cv=5, scoring='accuracy', n_jobs=-1)

# Fit model ke data
grid_search_rf.fit(X_train, y_train)

# Menampilkan hasil terbaik
print(f"Best parameters for Random Forest: {grid_search_rf.best_params_}")
print(f"Best cross-validation accuracy for Random Forest: {grid_search_rf.best_score_}")
```

```python
import matplotlib.pyplot as plt

# Calculate accuracy for each model
from sklearn.metrics import accuracy_score

accuracy_linreg = accuracy_score(y_test, y_pred_logreg)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
accuracy_svm = accuracy_score(y_test, y_pred_svm)

# Model names and accuracies
models = ['Logistic Regression', 'Random Forest', 'SVM']
accuracies = [accuracy_linreg, accuracy_rf, accuracy_svm]

# Create bar plot
plt.figure(figsize=(10, 6))
bars=plt.bar(models, accuracies, color=['#cdb4db', '#ffafcc', '#a2d2ff'])
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.ylim(0, 1)

# Adding accuracy values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval * 100:.2f}%', ha='center', va='bottom')

plt.show()
```
